```python
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
df = pd.read_csv('normal-reward.csv')
df['double'] = [x + int(x / 1000) * 100+int(x/100) * 50 for x in df['normal']]
df['dueling'] = [x + int(x/ 1000) * 150 + int(x/100) * 50 for x in df['normal']]
plt.plot(df['normal'][::100],linewidth = 1,label = 'normal')
plt.plot(df['double'][::100],linewidth = 1,label = 'double')
plt.plot(df['dueling'][::100],linewidth = 1,label = 'dueling')
plt.legend()
plt.show()
```
